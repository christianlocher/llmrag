# V.1 Database Request and Assistant Integration

```python
import os
import chromadb
from google.colab import ai
from dotenv import load_dotenv

load_dotenv()

# setting the environment


DATA_PATH = r"sample_data"
CHROMA_PATH = r"chroma_db"


chroma_client = chromadb.PersistentClient(path=CHROMA_PATH)

collection = chroma_client.get_or_create_collection(name="Furniture")

print("Start messaging with the bot (type quit to stop)!")

while True:

  user_query = input("How can I help you?")

  if user_query.lower() == "quit":
            break

  # Google Interface doese not support extra system prompt and user prompt ==> Use prompt engineering
  results = collection.query(
      query_texts=[user_query],
      n_results=10

  )

  system_prompt = """
  Du bist ein hilfreicher Recherche-Assistent. Du gibst so umfassende Antworten wie möglich. Deine Antworten basieren jedoch nur auf den Daten, die ich Dir in diesem Prompt mitsende. Wenn Du aufgrund der Daten nicht antworten kannst, antworte mit "In den Daten liegt kein Hinweis vor."
  --------------------
  The data:
  """+str(results['documents'])+"""
  """

  print(system_prompt)

  combined_prompt = f"""
  System Instructions:
  {system_prompt}

  User Query:
  {user_query}
"""


  response = ai.generate_text(combined_prompt, model_name='google/gemini-2.5-flash-lite')


  print("\n\n---------------------\n\n")

  print(response)



```
